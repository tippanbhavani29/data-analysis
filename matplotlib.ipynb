# Matplotlib -Python visualization library

## What is Matplotlib?
    Matplotlib is a widely used python library used for creating static ,animated and interactive data visualizations.
## Why it is need it?
    Helps to understand trends , patterns , and outlayers
## Where we use it?
    1.Data Analysis
    2.Machine learning 
    3.Business purpose(to analyse the market trend , finance and scienific research)


```python
##importing matpotlib
import matplotlib.pyplot as plt
```

```python
##line plot
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
##Bar plot
catergories=['A','B','C','D']
values=[10,15,7,12]
plt.bar(catergories,values,color='blue')
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
##Scatter plot
x=[5,6,3,2,6,8,8,6,7]
y=[99,29,89,94,93,92,84,95,87]
plt.scatter(x,y,color='blue',marker='o')
plt.title("Scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
##histogram
data=[22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
plt.hist(data,bins=3,color='green',edgecolor='yellow')
plt.title("Histogram plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# customization
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Line 1')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend() #line info
plt.show()
```

```python
# subplots

x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
 
plt.subplot(1,2,1)  # 1 row, 2 columns, plot 1
plt.plot(x, y1, color='blue')
plt.title("Plot 1")
 
plt.subplot(1,2,2)  # 1 row, 2 columns, plot 2
plt.plot(x, y2, color='red')
plt.title("Plot 2")
 
plt.tight_layout()
plt.show()
```

```python
# save plots 
plt.plot(x, y1)
plt.title("Save Plot Example")
plt.savefig("my_plot.png")  # Saves in current directory
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
 
plt.pie(sizes , labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
 
```

```python
# box plot - to find outlier visually
data = [25, 30, 35, 35, 40, 45, 50, 55, 60]
plt.boxplot(data)
plt.title("Box Plot Example (Statistical Distribution)")
plt.ylabel("Values")
plt.show()
```

```python
#Violin plot
# visualize the distribution, probability density, and summary statistics (like median, quartiles) of numeric data across categories.
import numpy as np
data=np.random.normal(0,1,100)
plt.violinplot(data)
plt.title("Violin plot")
plt.show()
```

```python
##stem plot
## is a way to visualize discrete data points
x=np.linspace(0,2*np.pi,10)
y=np.cos(x)
plt.stem(x,y)
plt.title("Stem plot")
plt.show()
```

```python
##step plot
##is a type of line chart that connects data points with horizontal and vertical lines
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.step(x,y,where='mid',color='magenta')
plt.title("step plot")
plt.show()

```

```python
##Area plot 
#An area plot displays data as filled regions under a line, helping you visualize both the trend and the magnitude of change.
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.fill_between(x,y,color='lightblue',alpha=0.6)
plt.title("Area plot")
plt.show()
```

```python
##contour plot
##is a graphical representation that shows how a 3D surface varies across 2D using contour lines
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-3,3,30)
y=np.linspace(-3,3,30)
X,Y=np.meshgrid(x,y)
Z=np.sqrt(x**2+Y**2)
plt.contour(X,Y,Z,cmap='plasma')
plt.colorbar()
plt.show()
```

```python
plt.contourf(X,Y,Z,cmap='viridis')
plt.title("Filled contour")
plt.colorbar()
plt.show()
```

```python
##polar plot
## is type of a graph where data is reperesented in terms of angles and raddi rather than cartesian coordinates(x,y)
theta=np.linspace(0,2*np.pi,100)
r=np.abs(np.sin(2*theta))
plt.polar(theta,r)
plt.title("Polar plot")
plt.show()
```

```python
# errorr bar plot
##Displays uncertainty or variation in data 

x = np.arange(1,6)
y = np.random.randint(5,15,5)
errors = np.random.rand(5)
 
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot Example")
plt.show()
```

```python
# horizontal bar plot

categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.show()
```

```python
#stack plot 
#shows multiple datasets stacked on top of each other
days = [1,2,3,4,5]
eating = [2,3,4,3,4]
sleeping = [8,7,8,6,7]
working = [8,9,7,8,8]
 
plt.stackplot(days, eating, sleeping, working, labels=['Eating','Sleeping','Working'])
plt.legend()
plt.title("Stack Plot Example")
plt.show()
```

```python
#hexbin plot
#is used to visualize the relationship between two numerical variables when you have a large number of data points 
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=25, cmap='Blues')
plt.colorbar()
plt.title("Hexbin Plot Example")
plt.show()
```

```python
#3D line plot 
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
t=np.linspace(0,10,100)
x=np.sin(t)
y=np.cos(t)
z=t
ax.plot3D(x,y,z,'purple')
ax.set_title("3D line plot")
plt.show()
```

```python
## 3D scatter plot 
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
ax.scatter(x,y,z,color='red',marker='o')
ax.set_title("3D scatter")
plt.show()
```

```python
# wireframe plot
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title("3D Wireframe Plot")
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_title("3D Contour Plot")
plt.show()
```

```python
#3D bar plot

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = [5, 2, 7, 3, 6]
 
ax.bar3d(x, y, z, dx, dy, dz, color='teal')
ax.set_title("3D Bar Plot")
plt.show()
```

```python
#3D surface 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
 
 
```

```python
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
 
```

```python
#.Customize a plot with different colors, markers, and line styles.

x = [1, 2, 3, 4]
y1 = [10, 20, 15, 25]
y2 = [25, 15, 20, 10]
plt.figure(figsize=(8, 5))
plt.plot(x, y1, color='red', marker='o', linestyle='--')
plt.plot(x, y2, color='green', marker='s', linestyle='-.')
plt.title('Customized Plot Example')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(['Line 1', 'Line 2'])
plt.show()
```
