# Seaborn - statistical data visualization

## what is seaborn?
    seaborn is a python library for creating statistical visualizations .built on matplotlib and integrated with pandas ,that makes the statistical data more easy and beautiful

## Why we need it?
    1.For creating plots like violin ,boxplot . it simplifies complex plots with  minimal code
    2.it provide prebuilt themes and statistical plots that turn complex data to visuals
    3.Automatically handles data from pandas DataFrames 
## Where we use it?
    1.Data analysis and reporting 
    2.Exploring relations between variables
    3.Machine learning (for visalizing dataset and result)

```python
#importing 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load dataset
df =sns.load_dataset('titanic')
df.head()
```

```python
df.shape
```

```python
##Histograms & countplots
sns.histplot(df['age'].dropna(),bins=10,kde=True)
plt.title("Histogram")
plt.show()

sns.countplot(x='class',data=df)
plt.title("Count of a Passenger by class")
plt.show()
```

```python
##Scatter plot 
sns.scatterplot(x='age',y='fare',data=df,hue='sex')
plt.title("Age Vs Fare")
plt.show()
```

```python
### Boxplot
sns.barplot(x='class',y='age', data=df)
plt.title("Boxplot of Age by class")
plt.show()
```

```python
##violin plot
sns.violinplot(x='class',y='age',hue='sex',data=df,split=True)
plt.title("Violin plot")
plt.show()
```

```python
#Pair plot
iris=sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
plt.show()
```

```python
## heatmap
##correlation heatmap 
corr=iris.select_dtypes(include=['Float64','int64']).corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')

plt.title("Coorelation heat map")
plt.show()
```

```python
##Customization and themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python

sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
 
```

```python
"Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```

```python
# 1. Compare Passenger Survival by Class and Gender
sns.barplot(x='class', y='survived', hue='sex', data=df)
plt.title("Passenger Survival Rate by Class and Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()
```

```python
# 2. Distribution of Passenger Ages
sns.histplot(df['age'].dropna(), bins=20, kde=True, color='skyblue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# --- 3. Fare Distribution by Class ---
plt.figure(figsize=(8, 5))
sns.boxplot(x='class', y='fare', data=titanic, palette='pastel')
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
#4 Relation between age and fare
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='fare', hue='class', data=titanic, palette='coolwarm', alpha=0.7)
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#5 Corrrelation heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
plt.figure(figsize=(8, 6))
corr = titanic.corr(numeric_only=True)   # Include only numeric columns
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
