# Pandas -Data Analysis Library
## 1.What is pandas?
    Pandas is a python libraru for data manipulation and analysis.It provides two main data structures:
        ->series:-1-dimensional labeled array(like a column)
        ->DataFrame:-2-dimensional tabular data(like a spreadsheet or SQL table)
## Why Learn Pandas?
    Pandas is essential because it simplifies working with data:
    1.Data Cleaning and Exploration:Easily handles missing values,filter data ,and summarize datasets.
    2.Efficient Handling of large datasets:Optimized for performance ,built on top of numpy
    3.Data Manipulation:Merge,join,pivot,group,and reshape datasets with simple commands

## Where is Pandas Used?

    Data Analysis & Reporting – Summarize, visualize, and report insights from data.

    Machine Learning – Preprocess and prepare datasets for ML models.

    Domain-specific Applications – Finance, healthcare, marketing, and more.

    Data Handling – Work seamlessly with CSV, Excel, SQL databases, and other file formats.

```python
#importing pandas
import pandas as pd

```

```python
 #3.Series & DataFrames
# Series
s = pd.Series([10, 20, 30, 40]) 
print("Series:\n", s) 
# DataFrame from dict
data = {'Name': ['Alice','Bob','Charlie'], 'Age':[25,30,35]} 
df = pd.DataFrame(data) 
print("\nDataFrame:\n", df)
```

```python
# 4.Reading and exploring the data

# Read CSV (replace with your file path)
 #df = pd.read_csv("titanic.csv")
 
# Example DataFrame
data = {'Name':['Alice','Bob','Charlie'],'Age':[25,30,35],'Sex':['F','M','M']}
df = pd.DataFrame(data)
 
# Explore
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())#statistics
print(df.shape)#no .of rowsand col
```

```python
#5. Selecting Columns & Rows
# Select column
print(df['Name'])
 
# Select multiple columns
print(df[['Name','Age']])
 
# Select row by index
print(df.iloc[0])       # first row
print(df.loc[0])        # first row using label
 
# Select subset of rows and columns
print(df.loc[0:1, ['Name','Sex']])
```

```python
#6. Filtering & Sorting
# Filter rows
print(df[df['Age'] > 28])
 
# Sort by Age
print(df.sort_values('Age', ascending=False))
```

```python
#7. Adding & Removing Columns
# Add new column
df['Age_in_5yrs'] = df['Age'] + 5
print(df)
 
# Drop column
df = df.drop('Age_in_5yrs', axis=1)
print(df)
```

```python
#8. GroupBy & Aggregation
data = {'Name':['Alice','Bob','Charlie','Alice','Bob'],
        'Score':[85,90,95,80,70]}
df = pd.DataFrame(data)
 
# Group by Name and calculate mean score
grouped = df.groupby('Name').mean()
print(grouped)
```

```python
#9. Handling Missing Data
data = {'Name':['Alice','Bob','Charlie','David'],
        'Age':[25, None, 35, 40]}
df = pd.DataFrame(data)
 
# Fill missing value
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
 
# Drop rows with missing values
df = df.dropna()
```
