### Univariate Analysis 

```python
# Distribution of numerical columns
numeric_cols = ['Unit price', 'Quantity', 'Tax 5%', 'Sales', 'cogs', 'gross income', 'Rating']
df[numeric_cols].hist(figsize=(15,10), bins=20)
plt.show()

# Count plots for categorical columns
categorical_cols = ['Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Payment']
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### Bivariate Analysis 

```python
# Sales vs Product line
plt.figure(figsize=(10,5))
sns.boxplot(x='Product line', y='Sales', data=df)
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,7))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.show()
```

### Outlier treat


```python
# Example for 'Sales'
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['Sales'] < lower) | (df['Sales'] > upper)]
print("Number of outliers in Sales:", outliers.shape[0])

df['sales_capped']=df['Sales'].clip(lower,upper)
sns.boxplot(x=df['sales_capped'])
plt.title("Sales after outlier capping")
plt.show()
# Option 1: Remove outliers
df_cleaned = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]
```

### Feature Engineering 

```python

# Convert 'Date' column to datetime
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], errors='coerce')

# Extract Day, Month, Hour
df_cleaned['Day'] = df_cleaned['Date'].dt.day
df_cleaned['Month'] = df_cleaned['Date'].dt.month
df_cleaned['Hour'] = pd.to_datetime(df_cleaned['Time'].astype(str)).dt.hour

# Create Profit Margin
df_cleaned['Profit_Margin'] = df_cleaned['gross income'] / df_cleaned['cogs']
```
